```python
plt.rcParams['figure.figsize'] = (10.0, 3.0) # set default size of plots
# "Names" of classes:
class_names = ['loafs', 'pugs']
num_classes = len(class_names)

samples_per_class = 7 # Number of images from each class we want to see
for y, cls in enumerate(class_names):
    idxs = np.flatnonzero(labels == y)
    idxs = np.random.choice(idxs, samples_per_class, replace=False)
    for i, idx in enumerate(idxs):
        plt_idx = i + samples_per_class * y + 1
        plt.subplot(num_classes, samples_per_class, plt_idx)
        plt.imshow(images[idx], plt.get_cmap('gray'))
        plt.axis('off')
        if i == 0:
            plt.title(cls)
plt.tight_layout()
plt.show()
```

```python
# Test train split happens here...
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size = 0.33, random_state = 11)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

Normalization step:
0-255 (they are 0-1)
the same format with the inicial pug, so the same methods can be applied
to make it better for histogram approach

```python
normalise = lambda image: ((image + np.abs(np.min(image)))/(np.max(image + np.abs(np.min(image)))) * 255).astype('uint8')
X_train = np.array([normalise(image) for image in X_train])
X_test = np.array([normalise(image) for image in X_test])
```

## Classical CV algorithm based on histogram of pixel intensity

We will assume that **pugs and loafs have different distributions of pixel intensities**. This assumption should help us distinguish between loafs and pugs.

```python
# Load images
loaf = X_train[np.flatnonzero(y_train == 0)[3]]
pug = X_train[np.flatnonzero(y_train == 1)[5]]

# Plot images
plt.subplot(1, 2, 1)
plt.title('Loaf')
plt.axis('off')
plt.imshow(loaf)

plt.subplot(1, 2, 2)
plt.title('Pug')
plt.axis('off')
plt.imshow(pug)
plt.tight_layout()
```

Here we test our assumption and visualise the distribution of pixel intensities for the above two images.

```python
plt.rcParams['figure.figsize'] = (10.0, 3.0)

# We will start by taking average of a pixel's RGB channels.
# Then we flatten the result into a 1D array.
loaf_mean = loaf.mean(axis=2).flatten()

# Now for pug
pug_mean = pug.mean(axis=2).flatten()

# Calculate and visualise
plt.subplot(1, 2, 1)
loaf_hist = plt.hist(loaf_mean, bins = range(256))
(loaf_counts, loaf_bins, _) = loaf_hist
plt.xlim([-0.5, 255.5])
plt.title('Loaf Histogram')

plt.subplot(1, 2, 2)
pug_hist = plt.hist(pug_mean, bins = range(256))
(pug_counts, pug_bins, _) = pug_hist
plt.title('Pug Histogram')
plt.xlim([-0.5, 255.5])
plt.show()
```

## Train a CNN model

Let's train a CNN model to do the same thing. Here we will use Keras (with TensorFlow on the background), as we did in the Machine Learning course. Further in the course, we will switch to PyTorch, as this seems to be a preferred framework when it comes to cutting edge research and technology.

Below you might recognise a CNN model that we have used in the Machine Learning course. In fact, this is almost a replica for the LeNet model proposed by Yann LeCun many years ago.

```python
from tensorflow.keras.layers import Input, Conv2D, Activation, Flatten, Dense, MaxPooling2D, BatchNormalization, Dropout

x = Input(shape=(224, 224, 3))
c1 = Conv2D(32, (3, 3), strides=(1, 1), padding="same")(x)
b1 = BatchNormalization()(c1)
a1 = Activation('relu')(b1)
c2 = Conv2D(32, (3, 3), strides=(1, 1), padding="valid")(a1)
b2 = BatchNormalization()(c2)
a2 = Activation('relu')(b2)
p2 = MaxPooling2D(pool_size=2)(a2)
d2 = Dropout(rate=0.50)(p2)
f2 = Flatten()(d2) # tensor into vector
h3 = Dense(100)(f2)
b3 = BatchNormalization()(h3)
a3 = Activation('relu')(b3)
d3 = Dropout(rate=0.5)(a3)
z = Dense(1)(d3)
p = Activation('sigmoid')(z)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

model = Model(inputs=x, outputs=p)
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])
model.summary()
```

```python
# Fit the model; returns History object
history = model.fit(X_train, y_train, batch_size=64, epochs=10, validation_data=(X_test, y_test))
```

What if we take a bigger and more established model such as ResNet? After the Alexnet, ResNet was one of the most important breakthrough in the field. Because of its novel residual connections that allowed data to skip certain parts of the model, researchers were able to build ever deeper models without loosing the representational power of convolutional layers. Read more about ResNet: https://builtin.com/artificial-intelligence/resnet-architecture.

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from keras.layers import Dense, GlobalAveragePooling2D, Input

# here we are simply loading ResNet50 (50 means the number of residual blocks)
base_model = ResNet50(weights=None, include_top=False)

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
predictions = Dense(1, activation='sigmoid')(x)
model = Model(inputs=base_model.input, outputs=predictions)
for layer in base_model.layers:
    layer.trainable = False
```

```python
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, batch_size=64, epochs=10, validation_data=(X_test, y_test))
```

What if we use model which was pretrained on a large ILSVRC dataset instead of training it from scratch? By the way, using pre-trained models to new datasets is called **transfer learning**.

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from keras.layers import Dense, GlobalAveragePooling2D, Input

base_model = ResNet50(weights='imagenet', include_top=False)

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
predictions = Dense(1, activation='sigmoid')(x)
model = Model(inputs=base_model.input, outputs=predictions)
for layer in base_model.layers:
    layer.trainable = False
```

```python
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, batch_size=64, epochs=10, validation_data=(X_test, y_test))
```
